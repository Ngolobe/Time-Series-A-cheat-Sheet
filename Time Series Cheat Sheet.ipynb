```python
import pandas as pd

# package to extract data from various Internet sources into a DataFrame
# make sure you have it installed
from pandas_datareader import data, wb

# package for dates
import datetime as dt
```

# Create your time range (start and end variables). The start date should be 01/01/2015 and the end should today (whatever your today is)

```python
start = dt.datetime(2015, 1, 1)
end = dt.datetime.today()
start
end
```

#  Select the Apple, Tesla, Twitter, IBM, LinkedIn stocks symbols and assign them to a variable called stocks



```python
#eg:apple =["AAPL"]

stocks = ['AAPL', 'TSLA','TWTR', 'IBM', 'LNKD']
```

# Read the data from yahoo, assign to df and print it

```python
import pandas_datareader as web
df = web.DataReader(stocks, 'yahoo', start, end)
df
```

 # What is the type of structure of df ?

```python
df.info(verbose=False)

print('\n','pandas.core.frame.DataFrame')
```

# Create a dataFrame called vol, with the Volume values.

```python
vol = df['Volume']
vol.head()
```

 # Aggregate the data of Volume to weekly

```python
vol['week'] = vol.index.week
vol['year'] = vol.index.year

weekly_sum = vol.groupby(['week','year']).agg('sum')
weekly_sum
vol
```

 # Find all the volume traded in the year 

```python
yearly_vol_sum = vol.groupby(['year']).agg('sum')
yearly_vol_sum
```

 # Find all the volume traded in the year 2019

```python
vol.loc['2019',stocks].sum()
```

# Import the apple stock only


```python
df_apple = web.DataReader('AAPL', 'yahoo', start, end)
df_apple
```

 # Check out the type of the columns

```python
df_apple.info()
```

# set Date as columns

```python
# Reset the index of dataframe
modified = df_apple.reset_index()
modified
```

# Transform the Date column as a datetime type

```python
modified['Date'] = pd.to_datetime(modified['Date'])
modified
```

# Set the date as the index

```python
modified.set_index('Date', inplace=True)
modified.columns
```

 # Is there any duplicate dates?

```python
# modified.iloc[0]
modified.index.duplicated().sum()
```

 # Make the first entry the oldest date.

```python
#quite confusing but ...
modified.sort_index(ascending=True)
#modified.sort_index(ascending=False)
```

# Get the last business day of each month

```python
df_lastBD = modified.loc[modified.index.isin(modified.index + pd.offsets.BMonthEnd(1))]
# modified.index + pd.offsets.BMonthEnd(1)
df_lastBD
```

# What is the difference in days between the first day and the oldest

```python
print((modified.index.max() - modified.index.min()).days)
```

# How many months do we have in the data ?

```python
import numpy as np
print((modified.index.max() - modified.index.min())/np.timedelta64(1, 'M'),'\n', 'or')
df_lastBD.shape[0]
```
